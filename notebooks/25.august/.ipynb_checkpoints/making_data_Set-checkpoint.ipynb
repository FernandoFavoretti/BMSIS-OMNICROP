```python
def make_nan_chart(df, title):
    nans = df.isnull().sum()
    x_axis = list(nans.index)
    x_pos = np.arange(len(x_axis))
    y_axis = list(nans.values)
    y_percent = map(lambda y: (y*1.0/int(df.shape[0]))*100, y_axis)
    plt.bar(list(x_pos), list(y_percent), align='center', alpha=0.5)
    plt.xticks(x_pos, x_axis,  rotation='vertical')
    plt.ylabel('num of N/A') 
    plt.title(title)
    plt.show()
```

```python
make_nan_chart(df, "Nan")
```

```python
  
def init_awhere():
    DATE_FORMAT = '%Y-%m-%d'
    awhere = Awhere()
    return awhere
```

```python
def complete_data(awhere,data_title, lat, long, period, period_month,data_existent):
    global last_long
    global last_lat
    global last_periodo
    global last_mean
    global counter
    counter += 1
    awhere = init_awhere()
    if last_long == long and last_lat==lat:
        print(counter)
        return last_mean
    
    last_long = long
    last_lat = lat
    last_periodo = period
    if period == "Growing season":
        months = period_month.split("-")
        switcher ={
            "Jan.": '01',
            "Feb.": '02',
            "Mar.": '03',
            "Apr.": '04',
            "May": '05',
            "Jun.": '06',
            "Jul.": '07',
            "Aug.": '08',
            "Sep.": '09',
            "Oct.": '10',
            "Nov.": '11',
            "Dec.": '12'
        }
        initial = switcher[months[0]]
        final = switcher[months[1]]
        date_range_i = initial+"-"+'01'
        date_range_f = final+"-"+'30'
    else:
        date_range_i = '01-01'
        date_range_f = '12-31'

    if not math.isnan(data_existent) and data_existent > -99:
        print(counter)
        last_mean = data_existent
        return data_existent
    
    else:
        result = awhere.norms_call(lat, long, date_range_i, date_range_f, '2005', '2007')
        result = result["norms"]

    if data_title == "mean_temp":
        means = []
        for days in result:
            means.append((days["meanTemp"]["average"]))
            
        mean= sum(means)/len(means)
        print(mean)
        last_mean = mean
        print(counter)
        return mean
    
    if data_title == "humidity":
        means = []
        for days in result:
            means.append((days["maxHumidity"]["average"]))
            
        mean= sum(means)/len(means)
        print(mean)
        last_mean = mean
        print(counter)
        return mean

    if data_title == "solar":
        means = []
        for days in result:
            means.append((days["solar"]["average"]))
            
        mean= sum(means)/len(means)
        print(mean)
        last_mean = mean
        print(counter)
        return mean
    
            
        
        
```

```python
soil_df = pd.read_csv('soil_dataset.csv', encoding='latin1')
soil_df2 = soil_df[['Latitude', 'Longitude']]
list_of_possibles_lats = soil_df2.T.to_dict().values()
df_soil = soil_df
```

```python
from math import cos, asin, sqrt

def distance(lat1, lon1, lat2, lon2):
    p = 0.017453292519943295
    a = 0.5 - cos((lat2-lat1)*p)/2 + cos(lat1*p)*cos(lat2*p) * (1-cos((lon2-lon1)*p)) / 2
    return 12742 * asin(sqrt(a))

def closest(data, v):
    best =  min(data, key=lambda p: distance(v['latitude'],v['longitude'],p['Latitude'],p['Longitude']))
    menor_distancia = distance(v['latitude'],v['longitude'],best['Latitude'],best['Longitude'])
    
    if menor_distancia < 160: #100 milhas
        print(best, menor_distancia)
        return best
    else:
        return None



```

```python
def get_soil_data(df_soil, list_of_possibles_lats, lat, long, type_of_data):
    v = {'latitude': lat, 'longitude': long}
    best_cords = closest(list_of_possibles_lats, v)
    df = df_soil.loc[ (df_soil['Latitude'] == best_cords['Latitude']) &
                         (df_soil['Longitude'] == best_cords['Longitude'])]

    if type_of_data == "pH":
        return df['pH'].iloc[0]
    
    if type_of_data == 'soil_order':
        return df['Soil_Order '].iloc[0]
    
    if type_of_data == 'bicarbonate':
        return df['Bicarbonate_Pi'].iloc[0]
    
    if type_of_data == 'C':
        return df['Organic _C'].iloc[0]
    
    if type_of_data == 'N':
        return df['Total_N'].iloc[0]
    
```
